```python
import pandas as pd
from sklearn.model_selection import GroupShuffleSplit
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from PIL import Image
import torch
import os
import matplotlib.pyplot as plt

# Load CSV file
csv_file_path = './Data_Entry_2017_v2020.csv'
df = pd.read_csv(csv_file_path)

# Define class names
class_names = [
    'Atelectasis', 'Consolidation', 'Effusion', 'Infiltration', 'Mass',
    'Nodule', 'Pneumothorax', 'Pleural_Thickening', 'Hernia', 'Pneumonia',
    'Fibrosis', 'Edema', 'Cardiomegaly', 'Emphysema', 'No Finding'
]

# Get labels function
def get_labels(diseases):
    labels = torch.zeros(len(class_names))
    if diseases != 'No Finding':
        for label_name in diseases.split('|'):
            if label_name in class_names:
                labels[class_names.index(label_name)] = 1
    return labels

# Dataset class
class CXR14dataset(Dataset):
    def __init__(self, img_dirs, df, augment=None):
        self.img_dirs = img_dirs
        self.df = df
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize([0.5123], [0.2307])
        ])
        self.augment = augment

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_name = self.df.iloc[idx, 0]
        diseases = self.df.iloc[idx, 1]
        label = get_labels(diseases)

        image = None
        for img_dir in self.img_dirs:
            img_path = os.path.join(img_dir, img_name)
            if os.path.exists(img_path):
                image = Image.open(img_path).convert("RGB")
                break

        if image is None:
            raise FileNotFoundError(f"Image {img_name} not found in provided directories.")

        if self.augment:
            image = self.augment(image)
        else:
            image = self.transform(image)

        return image, label

import random

def balance_multi_label_dataset_to_minimum(data, class_names):
    """
    Balance a multi-label dataset such that all labels reach the minimum count of any label.

    Args:
        data (pd.DataFrame): The dataset containing image indices and their labels.
        class_names (list): List of all possible classes.

    Returns:
        pd.DataFrame: A balanced dataset.
    """
    # Add a column with the list of labels for each image
    data['Labels List'] = data['Finding Labels'].apply(lambda x: [label for label in x.split('|') if label in class_names])

    # Compute the number of images contributing to each label
    label_counts = {label: 0 for label in class_names}
    for labels in data['Labels List']:
        for label in labels:
            label_counts[label] += 1

    # Determine the target count as the minimum label count
    min_label_count = min(label_counts.values())

    # Initialize label counts and sampled indices
    sampled_label_counts = {label: 0 for label in class_names}
    sampled_indices = []

    # Iterate until all labels reach the minimum count
    while any(sampled_label_counts[label] < min_label_count for label in class_names):
        # Randomly sample an image index
        idx = random.choice(data.index)

        # Get the labels associated with this image
        image_labels = data.loc[idx, 'Labels List']

        # Check if adding this image will exceed the target for any label
        if all(sampled_label_counts[label] < min_label_count for label in image_labels):
            # Add the image to the sampled list
            sampled_indices.append(idx)

            # Increment the count for each label associated with the image
            for label in image_labels:
                sampled_label_counts[label] += 1

    # Return the balanced dataset
    return data.loc[sampled_indices]



# Create combined labels column for stratification
df['combined_labels'] = df['Finding Labels'].apply(
    lambda x: '|'.join(sorted([label for label in x.split('|') if label in class_names]))
)

# Split data into train and test sets
group_split = GroupShuffleSplit(test_size=0.2, n_splits=1, random_state=42)
train_indices, test_indices = next(group_split.split(df, groups=df['Patient ID']))
train_df = df.iloc[train_indices]
test_df = df.iloc[test_indices]

print(f"Train data before balancing: {train_df.shape}")
print(f"Test data before balancing: {test_df.shape}")

# Balance the train and test datasets
balanced_train_df = balance_multi_label_dataset_to_minimum(train_df, class_names)


print(f"Balanced train data: {balanced_train_df.shape}")
print(f"Balanced test data: {test_df.shape}")

# Image directories
img_dirs = ['./DATA/images']

# Create datasets and dataloaders
batch_size = 64
num_cores = os.cpu_count()

train_dataset = CXR14dataset(img_dirs=img_dirs, df=balanced_train_df)
test_dataset = CXR14dataset(img_dirs=img_dirs, df=test_df)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=num_cores, pin_memory=True, persistent_workers=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=num_cores, pin_memory=True, persistent_workers=True)

# Function to visualize class distribution
def plot_class_distribution(df, title):
    exploded_labels = df['Finding Labels'].str.split('|').explode().value_counts()
    plt.figure(figsize=(12, 6))
    exploded_labels.plot(kind='bar')
    plt.title(title)
    plt.xlabel('Classes')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Plot class distributions
print("Class distribution in the full dataset:")
plot_class_distribution(df, "Class Distribution in Full Dataset")

print("Class distribution in the balanced train set:")
plot_class_distribution(balanced_train_df, "Class Distribution in Balanced Train Set")

print("Class distribution in the balanced test set:")
plot_class_distribution(test_df, "Class Distribution in Balanced Test Set")
```
